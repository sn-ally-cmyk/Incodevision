***VISUALIZATIONS***

Histogram (Distribution)

```python
df.hist(figsize=(12,10))
plt.show()
```

Correlation Heatmap

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

Bar Chart (Categorical Insight)

```python
df['Admission Status'].value_counts().plot(kind='bar')
plt.title("Admission Status Count")
plt.xlabel("Category")
plt.ylabel("Count")
plt.show()
```

Boxplot (Outlier Detection)

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df)
plt.xticks(rotation=90)
plt.title("Boxplot for Outlier Detection")
plt.show()
```

Pairplot (Relationships)

```python
sns.pairplot(df)
plt.show()
```

Scatter Plot (Important Relationship)

```python
sns.scatterplot(x='Admission Test Score', y='High School Percentage', data=df)
plt.title("Admission Test Score vs High School Percentage")
plt.show()
```

### Key Insights from EDA

1. Most features follow a normal distribution with slight skewness.
2. Strong correlation observed between GRE score and admission chances.
3. Some features show outliers which may affect model performance.
4. Dataset is now clean with no missing values.
5. Certain features have higher influence on target variable.
